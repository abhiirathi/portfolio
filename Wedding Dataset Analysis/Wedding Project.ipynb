```python
import pandas as pd #importing pandas library to create dataframes and perform the required operations

file = "wedding_cost_data.csv"

wedding_data = pd.read_csv(file) #reading the csv file
wedding_data.head()
```

```python
#the budget level for the wedding (i.e., an aggregation of vendor budget levels)
vendor_budget_summary = wedding_data.groupby('vendor_name')['subtotal'].sum().sort_values(ascending=False)
vendor_budget_summary
```

```python
#the product/service level data
product_service_data = wedding_data[['item_name', 'price_per_item', 'quantity', 'subtotal']].sort_values(by='subtotal', ascending=False)
product_service_data
```

```python
#department-level price summary (i.e., a department-level breakdown)
department_level_breakdown = wedding_data.groupby('functions')['subtotal'].sum().sort_values(ascending = False)
department_level_breakdown
```

```python
#total price of the wedding
total_price = wedding_data['subtotal'].sum()
total_price
```

```python
# Defining functions to format each section of the wedding cost breakdown

def format_vendor_budget_summary(vendor_budget):
    formatted_summary = "Vendor Budget Summary (prices in USD): \n\n" + \
                        "\n".join([f"{vendor}: {subtotal}" for vendor, subtotal in vendor_budget.items()])
    return formatted_summary

def format_product_service_data(product_service):
    formatted_data = "Product/Service Level Data (prices in USD): \n\n" + product_service.to_string(index=False)
    return formatted_data

def format_department_level_breakdown(department_breakdown):
    formatted_breakdown = "Department-Level Price Summary (prices in USD): \n\n" + \
                          "\n".join([f"{department}: {subtotal}" for department, subtotal in department_breakdown.items()])
    return formatted_breakdown

def format_total_price(total):
    return f"Total Price for the Wedding: ${total}"

# Applying the formatting functions to create the wedding cost breakdown

formatted_vendor_budget_summary = format_vendor_budget_summary(vendor_budget_summary)
formatted_product_service_data = format_product_service_data(product_service_data)
formatted_department_level_breakdown = format_department_level_breakdown(department_level_breakdown)
formatted_total_price = format_total_price(total_price)

# Combining all parts into a final output string
formatted_wedding_cost_breakdown = (
    "Wedding Cost Breakdown\n\n"
    f"{formatted_vendor_budget_summary}\n\n"
    f"{formatted_product_service_data}\n\n"
    f"{formatted_department_level_breakdown}\n\n"
    f"{formatted_total_price}\n"
)

print(formatted_wedding_cost_breakdown)
```
