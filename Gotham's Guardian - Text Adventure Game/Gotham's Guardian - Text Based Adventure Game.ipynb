```python
import random #Using it for random number generator
import time #Using it for time.sleep ; time.delay
import threading #Using it for the countdown in the code #Out of scope for this course, 
#but very essential element of the game to control for the time limit of 3 minutes. 
#So, decided to implement this
import sys # Import sys module to use sys.exit()

#Docstrings have been added for the functions in the code

def dark_knight_duel(available_stages):
    """
    Dark Knight Duel function for Batman's battle with Joker/Bane.
    Manages the game flow, interactions, and outcomes based on the player's choices.

    Parameters:
    available_stages (dict): A dictionary of stages available for the player to choose from.
    """
    
    # Character action handler
    def character_action(action, health, attack, dodge_chance, gadget_usage):
        """
        Modifies the character's stats based on the specified action.
        Supports various actions like healing, taking damage, adjusting attack strength,
        modifying dodge chance, and using gadgets.

        Parameters:
        action (str): The action to be performed (e.g., 'heal', 'take_damage').
        health (int): The current health of the character.
        attack (int): The current attack strength of the character.
        dodge_chance (float): The current dodge chance of the character.
        gadget_usage (int): The current count of gadgets available for use.

        Returns:
        tuple: Updated character stats (health, attack, dodge_chance, gadget_usage).
        """
        if action == "heal":
            health += 10
        elif action == "take_damage":
            health -= 10
        elif action == "weaken_attack":
            attack -= 10
        elif action == "strengthen_attack":
            attack += 10
        elif action == "increase_dodge_chance":
            dodge_chance += 0.1
        elif action == "decrease_dodge_chance":
            dodge_chance -= 0.1
        elif action == "use_gadget":
            gadget_usage -= 1
        return health, attack, dodge_chance, gadget_usage

    def display_bold(text):
        # Displays the given text in bold format for emphasis.
        BOLD = '\033[1m'  # Start of bold formatting.
        END = '\033[0m'   # End of bold formatting.
        print(BOLD + text + END)  # Print the text in bold.
        
    # Batman's action menu
    def batman_menu():
        """
        Displays Batman's action menu to the player.
        This menu lists all possible actions the player can take during their turn.
        """

        print("""
        Choose your action!
        
        Action | Rules
        1. Attack | Lower Batman's attack with each use. You start with 50 attack
        2. Use Gadget | 2 gadgets available; chance-based effectiveness.
        3. Dodge | Chance to avoid damage. Failure leads to taking damage.
        4. Heal | Heal Batman if health < 30; chance-based.
        5. Check Batman's Stats
        6. Check Villain's Stats
        7. Quit
    
        Key Points:
        Villain acts after Batman's turn.
        Win by depleting villain's health to 0; lose if Batman's health hits 0.
        """)
    
    #Took feedback from you in class and also added dialogues for the characters to make the combat game
    #feel more storylike for the user.
    
    # Define taunts for each villain
    joker_taunts = [
        "Joker: 'Come on, Batman, show me what you've got!'",
        "Joker: 'Let's put a smile on that face!'",
        "Joker: 'You can't beat me, Batman!'",
        "Joker: 'Why so serious? Let's dance, Batman!'",
        "Joker: 'This city deserves a better class of criminal, and I'm gonna give it to them!'"
        # Add more taunts here
    ]

    bane_taunts = [
        "Bane: 'You merely adopted the dark. I was born in it, molded by it.'",
        "Bane: 'I will break you, Batman!'",
        "Bane: 'Gotham will fall!'",
        "Bane: 'Your spirit, or your body: which shall I break first?'",
        "Bane: 'Gotham's reckoning has come at my hands.'"
        # Add more taunts here
    ]
    
    # Define Batman's taunts for dodging
    batman_taunts_on_dodge = [
    "Batman: 'Is that all you've got?'",
    "Batman: 'You'll have to do better than that!'",
    "Batman: 'Too slow!'",
    "Batman: 'Not good enough!'",
    "Batman: 'I'm always one step ahead.'"
    ] # Add more taunts here 
    
    # Game introduction
    display_bold("Welcome to Gotham City")
    time.sleep(2) # Pauses the game for 2 seconds for dramatic effect
    display_bold("Your mission as Batman is to defend Gotham from its notorious villains.")
    print("""
              _____       _____
     ,-'``_.-'` \   / `'-._``'-.
   ,`   .'      |`-'|      `.   `.
 ,`    (    /\  |   |  /\    )    `.
/       `--'  `-'   `-'  `--'       \
|                                   |
\      .--.  ,--.   ,--.  ,--.      /
 `.   (    \/ lt.\ /    \/    )   ,'
   `._ `--.___    V    ___.--' _,'
      `'----'`         `'----'` 
    
    """)
    time.sleep(2)

    # Selecting the villain
    while True:
        game_choice = input(prompt = """
        +------------------+-----------+
        |     Villain      |  Level    |
        +------------------+-----------+
        |      Joker       |  Hard     |
        |      Bane        |  Medium   |
        +------------------+-----------+
        Select your villain:
        Press 1 for Joker
        Press 2 for Bane
        """)
        # This input prompt asks the player to choose a villain to confront.
        # The player can select either Joker (hard level) or Bane (medium level),
        # or opt to return to the main menu.
        # Choosing Joker from The Dark Knight (my favourite villain of all time, mind-blowing acting by Heath Ledger)
        # Choosing Bane from The Dark Knight Rises
        
        # Setting up villain based on choice
        if game_choice == '1':
            villain_name = "Joker" 
            villain_health, villain_gadget_usage = 50, 2 # Joker's health and gadgets
            display_bold("You have decided to fight Joker.")
            time.sleep(1)
            print("\nJoker: 'Ah, Batman, we meet again. This time, Gotham's fate is in my hands!'")
            time.sleep(1)
            print("Batman: 'Your plans end here. Gotham is under my watch.'")
            time.sleep(1)
            break
        
        elif game_choice == '2':
            villain_name = "Bane"
            villain_health, villain_gadget_usage = 40, 1 # Bane's health and gadgets 
            # As Bane is medium difficulty, his health is reduced to 40 and also the gadgets are limited to 1 use per game
            display_bold("You have decided to fight Bane.")
            time.sleep(1)
            print("\nBane: 'Ah, Batman, we meet again. This time, Gotham's fate is in my hands!'")
            time.sleep(1)
            print("Batman: 'Your plans end here. Gotham is under my watch.'")
            time.sleep(1)
            break
        
        else:
            print("Invalid choice. Please select a villain to fight!")
            time.sleep(1)

    # Batman's initial stats
    # Initializing Batman's health, attack strength, dodge chance, and gadget usage
    # Played around with the Dodge chance for Batman, and initially thought of keeping it to 0.5
    # But, decided to make the game more favourable for the user by increasing the dodge probability to 0.6
    # Haven't done this in this version but for future updates, more gadgets can be introduced for batman 
    # like batarangs, smoke bombs, tasers, etc. and each gadget inflicts different damage on the villain 
    
    batman_health, batman_attack, batman_dodge_chance, batman_gadget_usage = 50, 50, 0.6, 2

    # Main game loop
    while batman_health > 0 and villain_health > 0:
        # Randomly select a taunt based on the villain
        if villain_name == "Joker":
            display_bold(random.choice(joker_taunts))
            time.sleep(1)
        elif villain_name == "Bane":
            display_bold(random.choice(bane_taunts))
            time.sleep(1)
        batman_menu() # Display the action menu for the player
        user_choice = input(prompt = "Select your move: ") # Player chooses an action 
        time.sleep(1)
        
        villain_action_taken = False # Flag to track if the villain has taken an action
        villain_dodge = False # Flag to track if the villain dodges an attack

        # Batman's actions
        
        if user_choice == "1":  # Player chooses to attack
            # Adjusted dodge chance for Joker (25% dodge chance) and Bane (~17% dodge chance)
            # Played around with the dodge chances too for the villains, had fun!
            # Again Bane's probability of dodging has been made less than Joker because the difficulty is Medium
            # Decrease Batman's attack strength with each attack
            
            if batman_attack <= 10:
                # If Batman's attack is too low, automatically call for backup
                print("Batman's attack is too low to be effective. Calling for backup!")
                time.sleep(1)
                # List of Batman's allies who can come as backup
                batman_allies = ["Robin", "Nightwing", "Batgirl", "Red Hood", "Alfred Pennyworth", "Commissioner Gordon", "Lucius Fox"]
                backup_helper = random.choice(batman_allies)  # Randomly select a backup helper
                print(f"{backup_helper} comes to support Batman! Attack increased.")
                
                # Strengthen Batman's attack
                batman_health, batman_attack, batman_dodge_chance, _ = character_action("strengthen_attack", batman_health, batman_attack, batman_dodge_chance, 0)
                time.sleep(1)
            else:
                # Decrease Batman's attack strength with each attack
                batman_health, batman_attack, batman_dodge_chance, _ = character_action("weaken_attack", batman_health, batman_attack, batman_dodge_chance, 0)
                villain_dodge_chance = random.randint(a=1, b=4) if villain_name == "Joker" else random.randint(a=1, b=6)
                if villain_dodge_chance == 1:
                    villain_dodge = True
                    print(f"{villain_name} dodges the attack!")  # Villain successfully dodges
                    time.sleep(1)
                else:
                 # Calculate if Batman's attack hits the villain
                    hit_chance = random.randint(a=0,b=9)
                    if (hit_chance < 7 and villain_name == "Joker") or (hit_chance < 8 and villain_name == "Bane"):  
                        # Batman's attack hits, and the villain's health is reduced
                        # More probability of hitting Bane
                        villain_health, _, _, _ = character_action("take_damage", villain_health, 0, 0, 0)
                        print(f"You hit {villain_name}! {villain_name}'s health: {villain_health}")
                        time.sleep(1)
                    else:
                        # Batman's attack misses
                        print("Your attack missed!")
                        time.sleep(1)
            villain_action_taken = True  # Mark that the villain has taken an action
            
        elif user_choice == "2":  # Use gadget
            if batman_gadget_usage > 0: # Check if Batman has gadgets left
            # Random chance to determine if the gadget use is successful
                success_chance = random.randint(a=0,b=9)
                if success_chance < 7:
                    # Gadget hits, villain's health is reduced
                    villain_health, _, _, _ = character_action("take_damage", villain_health, 0, 0, 0)
                    print(f"Your gadget worked! {villain_name}'s health: {villain_health}")
                    time.sleep(1)
                else:
                    # Gadget misses
                    print("The gadget missed its target!")
                    time.sleep(1)
                # Deduct a gadget from Batman's inventory
                batman_health, batman_attack, batman_dodge_chance, batman_gadget_usage = character_action("use_gadget", batman_health, batman_attack, batman_dodge_chance, batman_gadget_usage)
            else:
                # No gadgets left to use
                print("No more gadgets available.")
                time.sleep(1)
            villain_action_taken = True # Villain action marked as taken

        elif user_choice == '3':  # Dodge
            # Random chance to determine if Batman successfully dodges
            if random.random() < batman_dodge_chance:
                print("You successfully dodged the attack!")
                time.sleep(1)
                display_bold(random.choice(batman_taunts_on_dodge))  # Print a random taunt from Batman
                time.sleep(1)
                villain_action_taken = False  # Set villain action taken to False, indicating no attack from the villain this turn
            else:
                # Batman fails to dodge, takes damage
                print("You couldn't dodge in time.")
                time.sleep(1)
                batman_health, _, _, _ = character_action("take_damage", batman_health, 0, batman_dodge_chance, 0)
                print(f"{villain_name} hits Batman! Batman's health: {batman_health}")
                villain_action_taken = True
                time.sleep(1)
                
            # If batman fails to dodge the attack, the villain attacks for that turn automatically
            # and also gets an option to attack/ use gadget in the same round, thus having the 
            # chance to inflict double damage in a single turn
            # Implemented this to make it fair for villains too; we can't only favour Batman even though we love him!
                
        elif user_choice == '4':  # Heal
            if batman_health < 30:
                # If Batman's health is low, restore health
                display_bold("Attempting to recover Batman's critical health...")
                time.sleep(1)
                
                # Random chance for successful healing # 50-50 chance of healing/ not healing
                if random.choice([True, False]):
                    batman_health, batman_attack, batman_dodge_chance, _ = character_action("heal", batman_health, batman_attack, batman_dodge_chance, 0)
                    print(f"Batman's health has been restored! New health: {batman_health}")
                else:
                    print("Healing attempt failed! Batman must endure.")
                    time.sleep(1)
            else:
                # Healing not needed if Batman's health is greater than 30
                print("You don't need healing yet. Fight on!")
                time.sleep(1)
            villain_action_taken = True # Villain action marked as taken
            
        elif user_choice in ['5', '6']:  # Check stats
            if user_choice == '5': 
                # Display Batman's current stats
                print(f"Batman's Health: {batman_health}, Attack: {batman_attack}")
                time.sleep(1)
            elif user_choice == '6':
                # Display the villain's current health
                print(f"{villain_name}'s Health: {villain_health}")
                time.sleep(1)
            villain_action_taken = False # Checking stats doesn't count as an action

        elif user_choice == '7':  # Quit
            display_bold("Don't be a coward! Be a man, you decided to fight and now must play the game!")
            time.sleep(2)
            
            while True:  # Loop to ensure valid input
                quit_confirm = input("Still, if you really want to quit, this will count as a loss. Do you want to quit? (yes/no): ").lower()
        
                if quit_confirm == 'yes'.lower():
                    display_bold("Gotham needs its hero...\n")
                    time.sleep(1)
                    return "loss"  # This will signal that the player has chosen to quit, resulting in a loss.
        
                elif quit_confirm == 'no'.lower():
                    display_bold("Continuing the battle...\n")
                    time.sleep(1)
                    break  # Exit the loop and continue the game

                else:
                    print("Invalid input. Please enter 'yes' or 'no'.")
                
        # Villain's turn
        if villain_health > 0 and villain_action_taken:
            # Randomly choose the villain's action between attack and use gadget
            villain_action = random.choice(["attack", "use_gadget"])
            
            if villain_action == "attack":  
            # Set different hit chances for Joker and Bane
                villain_hit_chance = random.randint(a=0,b=9)
                hit_threshold = 7 if villain_name == "Bane" else 8  # Bane has a lower threshold to hit (medium difficulty), Joker has a higher threshold to hit(hard difficulty)

                if villain_hit_chance < hit_threshold:
                    # If the villain hits, reduce Batman's health
                    batman_health, _, _, _ = character_action("take_damage", batman_health, 0, batman_dodge_chance, 0)
                    print(f"{villain_name} attacks! Batman's health: {batman_health}")
                    time.sleep(1)
                else:
                    # If the villain misses, print a message indicating the miss
                    print(f"{villain_name}'s attack missed!")
                    time.sleep(1)
            elif villain_action == "use_gadget" and villain_gadget_usage > 0:
                # If the villain uses a gadget and has gadgets left, reduce Batman's health
                print(f"{villain_name} uses a gadget! Batman's health is reduced by 10!")
                batman_health, _, _, _ = character_action("take_damage", batman_health, 0, batman_dodge_chance, 0)
                time.sleep(1)
                # Deduct a gadget from the villain's inventory
                _, _, _, villain_gadget_usage = character_action("use_gadget", 0, 0, 0, villain_gadget_usage)

        # Check if Batman is defeated
        if batman_health <= 0:
            display_bold(f'Game Over! {villain_name} has defeated Batman.')
            time.sleep(1)
            display_bold("You lost this stage!\n")
            time.sleep(1)
            return "loss" # Return a loss for Batman
            break # Break out of the game loop

        # Check if the villain is defeated
        if villain_health <= 0: 
            display_bold(f'Congratulations! Batman has defeated {villain_name}.')
            time.sleep(1)
            display_bold("You won this stage!\n")
            time.sleep(1)
            return "win" # Return a win for Batman
            break # Break out of the game loop
            
#Future updates - 
# Introduce more villains with unique abilities and fighting styles to diversify challenges.
# Implement a more complex health system where damage and healing vary based on situation and gadget use.
# Add a range of new gadgets for Batman, each with distinct effects and usage limits.

def side_mission():
    """
    Conducts a side mission in the Batman game where the player needs to solve a series of riddles.
    The mission involves disabling a central control system to prevent a catastrophe in Gotham City.
    The player must answer riddles under time pressure, with limited hints available.
    """
    
    def display_bold(text):
        """
        Prints the given text in bold format.

        Parameters:
        text (str): The text to be printed in bold.
        """
        
        BOLD = '\033[1m'
        END = '\033[0m'
        print(BOLD + text + END)
        
    def display_rules():
        """
        Displays the rules of the side mission. 
        This includes the objective of disabling the central control system, 
        the consequence of failure, and the conditions for success.
        """
        
        display_bold("\nDisable the Central Control System")
        time.sleep(1)
        display_bold("Now, you are stuck and have to complete this stage. There's no going back.")
        time.sleep(1)
        display_bold("Central control system is in danger and if not stopped the bombs will explode and Gotham will be blown away!")
        time.sleep(2)
        print("Batman and Catwoman must solve riddles under time pressure.")
        time.sleep(1)
        print("Answer 3 out of 5 questions correctly within 60 seconds to disable the system.")
        time.sleep(1)
        print("You have 2 hints available to use.\n")
        time.sleep(1)

    def manage_countdown(time_limit, signal):
        """
        Manages the countdown timer for the side mission.

        Parameters:
        time_limit (int): The time limit for the mission in seconds.
        signal (dict): A dictionary used to communicate the state of the countdown 
                       (whether time is up or the countdown should stop).
        """
        while time_limit > 0 and not signal['stop']:
            print(f"\rTime remaining: {time_limit} seconds", end="")
            time.sleep(1)
            time_limit -= 1
        if not signal['stop']:
            signal['time_up'] = True
            display_bold("\nTime's up! The central control system activates. Gotham is in chaos.")
            display_bold("But, we still need an input to progess further in the story!")
            time.sleep(1)
    
    # A list of puzzles, each with a type (riddle, logic, pattern), a question, an answer, and a hint.
    # More puzzles can be added to this pool to increase the replay value of the game. 
    puzzles = [
        {"type": "riddle", "question": "What has keys but can't open locks?", "answer": "keyboard", "hint": "You use it to type."},
        {"type": "logic", "question": "What comes next? 2, 3, 5, 7, ?", "answer": "11", "hint": "Think of a series of prime numbers."},
        {"type": "pattern", "question": "What comes next in the sequence? 3, 8, 15, 24, ?", "answer": "35", "hint": "Each number increases by the next odd number."},
        {"type": "riddle", "question": "I’m light as a feather, yet the strongest person can't hold me for five minutes. What am I?", "answer": "breath", "hint": "It's something you do every moment."},
        {"type": "pattern", "question": "What comes next in the sequence? 1, 1, 2, 3, 5, ?", "answer": "8", "hint": "Each number is the sum of the previous two."},
        {"type": "riddle", "question": "I speak without a mouth and hear without ears. I have no body, but I come alive with the wind. What am I?", "answer": "echo", "hint": "I repeat what you say."},
        {"type": "pattern", "question": "What comes next in the sequence? 1, 2, 4, 8, ?", "answer": "16", "hint": "Each number is double the previous one."},
        {"type": "riddle", "question": "I'm tall when I'm young, and I'm short when I'm old. What am I?", "answer": "candle", "hint": "I light up your room and melt away."},
        {"type": "riddle", "question": "Forward I am heavy, but backward I am not. What am I?", "answer": "ton", "hint": "It's a unit of weight, but think in terms of letters."},
        {"type": "logic", "question": "What can travel around the world while staying in a corner?", "answer": "stamp", "hint": "It's something small and can be found on mail."},
        {"type": "riddle", "question": "The more you take, the more you leave behind. What am I?", "answer": "footsteps", "hint": "Think about what you create when you walk."},
        {"type": "logic", "question": "What word is spelled incorrectly in every dictionary?", "answer": "incorrectly", "hint": "Think about the question itself."},
        {"type": "riddle", "question": "What is always in front of you but can’t be seen?", "answer": "future", "hint": "It's a matter of time."},
        {"type": "logic", "question": "If you throw a red stone into the blue sea, what will it become?", "answer": "wet", "hint": "Think about what happens to objects in water."},
        {"type": "riddle", "question": "I am Batman's constant companion, younger and often seen in red. Who am I?", "answer": "Robin", "hint": "I am known as the Boy Wonder."},
        {"type": "logic", "question": "If the Batmobile is driving at 60 mph and doubles its speed every 10 minutes, how fast is it going after 30 minutes?", "answer": "480", "hint": "The speed doubles three times."}
    ]
    
    random.shuffle(puzzles) # Randomly shuffle the puzzles to ensure variety in each game session.
    
    correct_answers_needed = 3 # The number of correct answers required to successfully complete the mission.
    time_limit = 60 # The time limit for solving the puzzles, set to 60 seconds.
    puzzles_solved = 0 # Counter for the number of puzzles solved by the player.
    lifelines = 2 # The number of hints available to the player.
    signal = {'stop': False, 'time_up': False} # A dictionary to manage the state of the countdown and game.

    display_rules() # Display the rules of the side mission to the player.

    while True:
        # Continuously prompt the player to start the game until they respond with "yes".
        # Irritate him, the only response we would be accepting is Yes. This is a very crucial stage for the game and there's not going back. 
        user_response = input(prompt = "Should we start the game now? (yes): ").strip().lower()
        if user_response == "yes":
            break # Exit the loop if the player agrees to start the game
        else:
            display_bold("""You can only choose Yes here, no other choice! 
                  Gotham is in extreme danger of being blown away by the bombs!""")
    
    
    countdown_thread = threading.Thread(target=manage_countdown, args=(time_limit, signal))
    # Create a new thread for managing the countdown timer in the side mission.
    # The 'Thread' class is instantiated with the target function 'manage_countdown' and its arguments.
    # 'time_limit' specifies the duration of the countdown, and 'signal' is a dictionary used to control the countdown flow.
    
    countdown_thread.start()
    # Start the execution of the countdown thread.
    # This initiates the 'manage_countdown' function in a separate thread, allowing it to run concurrently with the main game logic.
    # This parallel execution ensures the countdown operates independently without interrupting or being interrupted by the main game flow.

    """NOTE: Explanation for Going Beyond the Curriculum with Threading:
    The inclusion of threading in this game, specifically for the countdown timer in the side mission, 
    is a strategic choice to enhance the game's realism and interactivity. By using a separate thread 
    for the countdown, the game simulates a real-time scenario where players must make decisions under time pressure.
    Also, this ensures that we stick to our time limit of 3 minutes and I have allocated 60 seconds for this stage"""

    for puzzle in puzzles[:5]:
        # Iterate through the first five puzzles in the shuffled list.
        if signal['time_up']:
            break # Break out of the loop if the countdown time is up.

        # Display the type and question of the current puzzle.
        display_bold(f"\n{puzzle['type'].title()} Puzzle: {puzzle['question']}")
        time.sleep(1)
        
        # Prompt the player to answer the puzzle or ask for a hint.
        if lifelines > 0:
            user_answer = input(prompt = "\nYour answer or type 'hint' for help: ").lower().strip()
        else:
            display_bold("You have run out of hints!")
            display_bold("Now, if you type 'hint' it will count as an incorrect answer.")
            user_answer = input(prompt = "\nYour answer: ").lower().strip()

        if signal['time_up']:
            break # Check again if time is up after the player's response.

        if user_answer == 'hint' and lifelines > 0:
            # If the player requests a hint and has lifelines remaining, show the hint.
            display_bold(f"\nHint: {puzzle['hint']}")
            lifelines -= 1 # Decrease the count of available lifelines.
            # Prompt the player to answer again after receiving the hint.
            user_answer = input(prompt = "Try again: ").lower().strip()

            if signal['time_up']:
                break # Check again if time is up after showing the hint.

        if user_answer == puzzle['answer'].lower():
            # If the player's answer matches the puzzle's answer, it's correct.
            print("\nCorrect! The security system is weakening.")
            puzzles_solved += 1 # Increment the count of puzzles solved.
            if puzzles_solved == correct_answers_needed:
                # If the required number of puzzles are solved, stop the countdown.
                signal['stop'] = True #Stop the countdown
                break # Break out of the loop as the mission is accomplished.
        else:
            # If the player's answer is incorrect, notify them.
            print("\nIncorrect answer. Be careful!")
            print(f"The correct answer was: {puzzle['answer']}")  # Display the correct answer

    signal['stop'] = True #Signal to the countdown thread to stop the countdown
    countdown_thread.join() #Joining the countdown thread

    """NOTE: The join() method ensures that the main program waits for the countdown thread to complete its tasks. 
    This synchronization is crucial for ensuring that the countdown fully stops and cleans up before the program proceeds, 
    preventing any potential conflicts or erratic behavior due to unfinished thread execution.
    """

    if signal['time_up']:
    # Check if the countdown has finished before the player solved enough puzzles.
        display_bold("Failed to disable the security. The central control system activates. Gotham is in peril.")
        display_bold("You have lost this stage!\n")
        return "fail"  # Return "fail" indicating that the player did not complete the mission in time.

    elif puzzles_solved >= correct_answers_needed:
    # Check if the player has solved the required number of puzzles within the time limit.
        display_bold("3 correct answers, great!")
        time.sleep(1)
        display_bold("Central system security disabled. Gotham City is safe!")
        display_bold("You have won this stage!\n")
    # Display a success message indicating the player successfully completed the mission.
        time.sleep(1)  
        return "success"  # Return "success" indicating the player has won the stage.

    else:
    # If the time hasn't run up but the player hasn't solved enough puzzles.
        display_bold("Failed to disable the security. The central control system activates. Gotham is in peril.")
        display_bold("You have lost this stage!")
    # Display a failure message indicating the player failed to complete the mission.
        time.sleep(1)  
        return "fail"  # Return "fail" indicating the player has lost the stage.

def gotham_challenge(available_stages):
    # Main function for the Gotham Challenge game mode.
    
    def typewriter_effect(text):
        # Simulates a typewriter effect when displaying text to enhance the user experience.
        for char in text:
            print(char, end='', flush=True)
            # 'char': Character to print.
            # 'end=''': Prevents new line after printing 'char', keeping subsequent prints on the same line.
            # 'flush=True': Ensures immediate display of 'char' by flushing the output buffer, crucial for real-time typewriter effect.
            time.sleep(0.03)  # with a slight delay between each character.
        print()  # Move to the next line after the text is fully printed.

    def display_bold(text):
        # Displays the given text in bold format for emphasis.
        BOLD = '\033[1m'  # Start of bold formatting.
        END = '\033[0m'   # End of bold formatting.
        print(BOLD + text + END)  # Print the text in bold.

    def display_rules():
        # Displays the rules and introduction for the Gotham Challenge.
        typewriter_effect("Batman faces Gotham's Strategic Challenge...")
        # Using the typewriter effect for a dramatic presentation of the challenge.
        typewriter_effect("""
        Gotham City faces a series of complex threats. As Batman, you need to use your detective skills 
        and strategic thinking to navigate these challenges and protect the city.
        """)
        display_bold("You must answer 3/5 correctly to win this stage!\n")
        time.sleep(1)  

    def start_game_prompt():
        """
    Repeatedly prompts the player to start the game, accepting only 'yes' or 'no' as valid responses.
    Returns True if the player confirms to start the game, otherwise continues prompting.

    Returns:
        bool: True if the player's response is 'yes', otherwise False.
    """
        while True:
            typewriter_effect("Should we start the game now? (yes/no): ")
            response = input().strip().lower()
            if response in ["yes", "no"]:
                return response == "yes"
            typewriter_effect("Invalid input. Please answer with 'yes' or 'no'.")

    def manage_countdown(time_limit, signal):
        """
    Manages a countdown timer for the game, decrementing from the specified time limit to zero.
    If 'time_up' in the signal dictionary is set to True externally, the countdown halts.
    When the time limit is reached without external interruption, 'time_up' is set to True,
    and a message indicating failure to save Gotham is displayed.

    Parameters:
        time_limit (int): The time limit for the countdown in seconds.
        signal (dict): A dictionary to control and track the countdown status.

    Returns:
        bool: False if the countdown reaches zero without external interruption.
    """
        for remaining in range(time_limit, 0, -1):
            if signal['time_up']:
                break
            print(f"\rTime remaining: {remaining} seconds", end="")
            time.sleep(1)
        if not signal['time_up']:
            signal['time_up'] = True
            display_bold("\nTime's up! You have failed this stage.") 
            display_bold("But, we still need an input to progess further in the story!")
            return "loss"
        
    # Scenarios would be defined here
    # More scenarios can be added to enhance the replay value of the game
    scenarios = [
        {"scenario": "Winter's Grip", "question": "I go up when rain comes down. What am I?", "choices": ["Umbrella", "Temperature"], "answer": "Umbrella"},
        {"scenario": "Green Enigma", "question": "I'm full of holes but still hold water. What am I?", "choices": ["Sponge", "Leaf"], "answer": "Sponge"},
        {"scenario": "Gotham's Secret", "question": "What belongs to you, but other people use it more than you?", "choices": ["Your name", "Your phone"], "answer": "Your name"},
        {"scenario": "Two-Face's Dilemma", "question": "What is so fragile that saying its name breaks it?", "choices": ["Silence", "Glass"], "answer": "Silence"},
        {"scenario": "Ra's al Ghul's Offer","question": "What was the name of the movie theater Bruce Wayne's parents were shot outside of?","choices": ["The Monarch Theater", "The Majestic Theater"],"answer": "The Monarch Theater"},
        {"scenario": "Man-Bat's Flight","question": "What is Batman's real first name?","choices": ["Ben", "Bruce"],"answer": "Bruce"},
        {"scenario": "Deathstroke's Ambush","question": "What tool does Batman use to scale tall buildings?","choices": ["Bat ladder", "Bat grappling hook"],"answer": "Bat grappling hook"},
        {"scenario": "Mad Hatter's Riddles","question": "Why can't a bicycle stand up by itself?","choices": ["It's two-tired", "It has no kickstand"],"answer": "It's two-tired"},
        {"scenario": "Robin's Riddle", "question": "What can you hold in your left hand but not in your right?", "choices": ["Your right elbow", "A secret"], "answer": "your right elbow"},
        {"scenario": "The Joker's Laughter", "question": "What begins with T, finishes with T, and has T in it?", "choices": ["A tournament", "A teapot"], "answer": "a teapot"},
        {"scenario": "Ra's al Ghul's Secret", "question": "Who trained Bruce Wayne in martial arts and deception?", "choices": ["Ra's al Ghul", "Alfred Pennyworth"], "answer": "Ra's al Ghul"},
        {"scenario": "Gotham's White Knight", "question": "Which district attorney becomes Two-Face?", "choices": ["Harvey Dent", "Carl Finch"], "answer": "Harvey Dent"},
        {"scenario": "Wayne's Legacy", "question": "What is the name of the corporate entity owned by Bruce Wayne?", "choices": ["Wayne Enterprises", "Gotham Enterprises"], "answer": "Wayne Enterprises"},
        {"scenario": "The Bat's Symbol", "question": "What does Batman use to summon help and cast fear into the heart of criminals?", "choices": ["The Bat-Signal", "The Bat-Phone"], "answer": "The Bat-Signal"},
        {"scenario": "The Bat's Cave", "question": "Where does Batman equip and strategize for his war on crime?", "choices": ["The Bat Bunker", "The Batcave"], "answer": "The Batcave"}
    ]
    
    random.shuffle(scenarios)
    # Randomly shuffles the list of scenarios to ensure a varied and unpredictable game experience each time.

    correct_answers_needed = 3
    # Sets the number of correct answers required for the player to successfully complete the challenge.

    superman_chase_helped = False
    # Initializes a flag to track whether Superman Chase has assisted the player in the game.

    answers_correct = 0
    # Initializes a counter for the number of correct answers given by the player.

    incorrect_answers = 0
    # Initializes a counter for the number of incorrect answers given by the player.

    time_limit = 60
    # Sets the time limit for the challenge to 60 seconds.

    signal = {'time_up': False}
    # Initializes a signal dictionary with a 'time_up' key to manage the countdown timer state.

    display_rules()
    # Calls the function to display the game rules with a typewriter effect.

    if not start_game_prompt():
        # Checks if the player is ready to start the game. If the player responds negatively,
        # the game outputs a message and exits the challenge.
        typewriter_effect("Exiting the challenge...")
        time.sleep(1)
        return

    #Same like done in the side mission challenge
    countdown_thread = threading.Thread(target=manage_countdown, args=(time_limit, signal))
    countdown_thread.start()

    for scenario in scenarios[:5]:
        # Iterates through the first five scenarios in the shuffled scenarios list.
        # This limits the game to a set number of scenarios per session, ensuring the total time does not exceed 3 minutes.
        
        if signal['time_up'] or answers_correct >= correct_answers_needed:
        # Checks if the countdown time is up or if the player has already answered enough questions correctly.
            break # Exits the loop if either condition is met.

        display_bold(f"\nScenario: {scenario['scenario']}")
        # Displays the title of the current scenario in bold for emphasis.

        display_bold(scenario['question'])
        # Displays the scenario's question in bold.

        for i, choice in enumerate(scenario['choices'], start=1):
        # Enumerates through the available choices in the current scenario, starting numbering from 1.
            print(f"{i}. {choice}")
        # Prints each choice with its corresponding number.

        while True:
        # A loop to get a valid response from the player.
            try:
            # Tries to get a numerical input from the player.
                choice_index = int(input(prompt = "Your decision (Please Enter 1 or 2): "))
                if choice_index in [1, 2]:
                # Checks if the input is 1 or 2, corresponding to the available choices.
                    choice = scenario['choices'][choice_index - 1].lower()
                # Assigns the player's choice based on their input.
                    break  # Exits the loop once a valid choice is made.
                else:
                # If the player enters a number other than 1 or 2.
                    print("Please choose either 1 or 2.")
            except ValueError:
            # Handles the case where the input is not a number.
                print("Invalid input. Please enter a number.")

            if signal['time_up']:
                return False # If time is up, exit the function immediately
            
        if signal['time_up']:
            # Check again after breaking out of the input loop
            return False

        # Check if the player's choice matches the correct answer for the scenario.
        if choice == scenario['answer'].lower():
            answers_correct += 1 # Increment the count of correct answers.
            # Check if the player still needs more correct answers to win.
            if answers_correct < correct_answers_needed:
                steps_remaining = correct_answers_needed - answers_correct
                # Inform the player how many more correct answers are needed to save Gotham.
                print(f"Correct! You're {steps_remaining} step{'s' if steps_remaining > 1 else ''} closer to saving Gotham.")
        else:
            incorrect_answers += 1 # Increment the count of incorrect answers.
            # Check if Superman Chase has not yet assisted.
            if not superman_chase_helped and not signal['time_up']:
                # Inform the player that Superman Chase is assisting and automatically provide a correct answer.
                display_bold("\nIncorrect answer! But why fear when Superman Chase is here.")
                #time.sleep(2) 
                #Thought of using time.sleep(2) here to add dramatic effect for 
                #Superman Chase's entry but it is counting in the timer and decided to not include it
                print("Superman Chase comes to your rescue and answers on your behalf!")
                print(f"By the way, the correct answer was: {scenario['answer']}\n")
                superman_chase_helped = True # Mark that Superman Chase has provided assistance.
                answers_correct += 1 # Increment the count of correct answers as Superman Chase helps.
            else:
                # If Superman Chase has already helped, inform the player of the incorrect answer and the risk to Gotham.
                print(f"Incorrect answer! The correct answer was: {scenario['answer']}\n")  
                # Display the correct answer

    signal['time_up'] = True
    # Set the 'time_up' signal to True to indicate the countdown is complete and the game stage should end.

    countdown_thread.join()
    # Wait for the countdown thread to finish its execution. 
    # This ensures that the countdown is fully completed before proceeding to evaluate the game outcome.

    # Evaluate the player's performance based on the number of correct answers given.
    if answers_correct >= correct_answers_needed:
        display_bold("\n Correct Answer! Congratulations, You have passed the stage.\n")
        return True # Return True to indicate a successful completion of the stage.

    else:  
        display_bold("\nAlas, some challenges remain unresolved. Gotham still needs its Dark Knight")
        display_bold("You lost this stage!\n")
        return False # Return False to indicate that the player did not successfully complete the stage.

#Superman Chase is our Easter Egg in this stage; surprise element to add fun to the game

def win():
    
    def display_bold(text):
        # Displays the given text in bold format for emphasis.
        BOLD = '\033[1m'  # Start of bold formatting.
        END = '\033[0m'   # End of bold formatting.
        print(BOLD + text + END)  # Print the text in bold.
    
    # Congratulatory message for winning the game
    display_bold("\nCongratulations! You have triumphed over Gotham's challenges and emerged victorious!")
    time.sleep(1)  # Short pause for dramatic effect
    display_bold("Gotham City is safe, thanks to Batman's bravery and skills. Well done Batman!")
    print("""
                    T\ T\
                | \| \
                |  |  :
           _____I__I  |
         .'            '.
       .'                '
       |   ..             '
       |  /__.            |
       :.' -'             |
      /__.                |
     /__, \               |
        |__\        _|    |
        :  '\     .'|     |
        |___|_,,,/  |     |    _..--.
     ,--_-   |     /'      \../ /  /\\
    ,'|_ I---|    7    ,,,_/ / ,  / _\\
  ,-- 7 \|  / ___..,,/   /  ,  ,_/   '-----.
 /   ,   \  |/  ,____,,,__,,__/            '\
,   ,     \__,,/                             |
| '.       _..---.._                         !.
! |      .' z_M__s. '.                        |
.:'      | (-_ _--')  :          L            !
.'.       '.  Y    _.'             \,         :
 .          '-----'                 !          .
 .           /  \                   .          .
    
    """)
    time.sleep(1)  # Another short pause

    # Loop for replay choice prompt
    while True:
        # Asking the player if they want to play again
        replay_choice = input("\nDo you want to play the game again? (yes/no): ").lower()
        if replay_choice == "yes":
            game_control()  # Restart the game if the player chooses to play again
            break  # Break out of the loop
        elif replay_choice == "no":
            print("Thank you for playing, Hope you had fun!")  # Farewell message
            sys.exit()  # Exit the game
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")  # Handle invalid inputs

def loss():
    
    def display_bold(text):
        # Displays the given text in bold format for emphasis.
        BOLD = '\033[1m'  # Start of bold formatting.
        END = '\033[0m'   # End of bold formatting.
        print(BOLD + text + END)  # Print the text in bold.
    
    # Message for losing the game
    display_bold("\nDespite your efforts, Gotham City falls into chaos.")
    time.sleep(1)  # Short pause
    display_bold("The villains have won this time, but Batman will return to fight another day.")
    time.sleep(1)  # Another short pause

    # Loop for replay choice prompt
    while True:
        # Asking the player if they want to play again
        replay_choice = input("\nDo you want to play the game again? (yes/no): ").lower()
        if replay_choice == "yes":
            game_control()  # Restart the game if the player chooses to play again
            break  # Break out of the loop
        elif replay_choice == "no":
            print("Thank you for playing! Goodbye!")  # Farewell message
            sys.exit()  # Exit the game
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")  # Handle invalid inputs

def game_control():
    """
    This is the main control function for the 'Gotham's Guardian' game. It orchestrates the game flow,
    keeps track of the player's progress across different stages, and handles the stage selection. The game
    includes three main stages, each with its unique challenges.

    The `available_stages` dictionary is used to dynamically track which stages are still available for the
    player to choose. Once a stage is completed, it's removed from this dictionary to prevent replaying
    the same stage within a single game session. This approach allows for a dynamic and flexible game flow.

    The `available_stages` argument is passed to other functions to ensure that each stage is aware of the
    overall game state and can modify the availability of stages accordingly. This inter-function
    communication is crucial for maintaining consistent game progression and state.
    """
    
    stage_wins = 0  # Tracks the number of stages won by the player
    stage_losses = 0  # Tracks the number of stages lost by the player
    
    # Dictionary of stages available for the player to choose from
    available_stages = {
        '1': 'Dark Knight Duel',
        '2': 'Side Mission with Catwoman',
        '3': 'Gotham Challenge'
    }
    
    def display_bold(text):
        # Displays the given text in bold format for emphasis.
        BOLD = '\033[1m'  # Start of bold formatting.
        END = '\033[0m'   # End of bold formatting.
        print(BOLD + text + END)  # Print the text in bold.
    
    # Introduction message for the game
    display_bold("Game Introduction")
    time.sleep(1)
    print("""
    Welcome to "Gotham's Guardian," a text-based adventure game where you step into the shoes of the 
    iconic superhero, Batman. Gotham City is under siege by its notorious villains, and it's up to you 
    to save the day. The game unfolds over three main stages, each presenting unique challenges 
    and decisions that shape your journey.

    In the Dark Knight Duel, you'll face off against either the cunning Joker or the powerful Bane, 
    using your skills, gadgets, and strategic thinking. The Side Mission with Catwoman takes you on a 
    thrilling race against time to solve riddles and disable a central control system. Finally, the 
    Gotham Challenge tests your detective skills and decision-making in critical scenarios to save the 
    city from chaos.

    As Batman, your actions and decisions will determine the fate of Gotham. Each stage offers a different
    gameplay experience, with elements of combat, puzzle-solving, and strategy. Be prepared to make tough 
    choices and face the consequences of your actions.
    """)
    time.sleep(2)  # Short pause for dramatic effect

    # Main loop for stage selection and progression
    while available_stages:
        print("Select a stage to begin:")
        for key, value in available_stages.items():
            print(f"{key}. {value}")

        choice = input(prompt="Enter your choice:")
        
        # Directing to the chosen stage
        if choice == '1':
            result = dark_knight_duel(available_stages)  # Starts the Dark Knight Duel
        elif choice == '2':
            result = side_mission()  # Starts the Side Mission with Catwoman
        elif choice == '3':
            result = gotham_challenge(available_stages)  # Starts the Gotham Challenge
        else:
            display_bold("Invalid choice. Please select from the available stages.\n")
            continue  # Continue the loop to prompt for stage selection again

        # Updating win/loss counts based on stage results
        if result in ["win", "success", True]:
            stage_wins += 1
        elif result in ["loss", "fail", False]:
            stage_losses += 1
        else:
            print("Returning to stage selection...\n")
            continue  # Returns to stage selection if no win/loss outcome

        # Determining the game's end based on win/loss count
        if stage_wins >= 2:
            win()  # Calls win function if player wins two or more stages
        elif stage_losses >= 2:
            loss()  # Calls loss function if player loses two or more stages

        # Updating available stages based on completion
        if result in ["win", "loss", "success", "fail", True, False]:
            del available_stages[choice]  # Removes completed stage
        else:
            print("Invalid choice or stage already completed, please select again.")

# Start the game
game_control()

# Had lots of fun coding, debugging, testing the game, great assignment! Came up with various scenarios 
# for the stages, lots of hit-and-trial approach. Now, its almost 8.50 PM, 3rd Dec and time to submit!
# Hope, you enjoy the game!
```
